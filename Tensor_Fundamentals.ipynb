### Creating a range of tensors and tensors like (takes the shape and creates those many zeroes or ones)

```python
one_to_thousand = torch.arange(start=1, end = 1000, step = 88)
one_to_thousand
```

```python
ten_zeros = torch.zeros_like(one_to_thousand)
ten_zeros
```

```python
ten_ones = torch.ones_like(one_to_thousand)
ten_ones
```
