### Reshaping, Viweing and Stacking Tensors

* Reshaoing - reshapes an input tensor to a defined shape
* View - Return a  view of an input tensor of certain shape but keep the same memory of the original tesnor
* Stacking - combine multiple tensors on top of each other or side by side
* Squeeze - removes all 1 dimensions froma a tensor
* Unsqueeze - add a 1 dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted(swapped) in a certain way

```python
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an  extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped
```

```python
# Change the view
z = x.view(3, 3)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original tensor)

z[:, 0] = 5
z, x
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([z, z, z, z, z])
x_stacked
```

```python
#torch.squeeze() - removes all single dimenstions from a target tensor
x_reshaped.squeeze(), x_reshaped.squeeze().shape
```

```python
x_reshaped, x_reshaped.shape
```

```python
print(f'x_shaped tensor shape {x_reshaped.shape}')
x_squeezed = x_reshaped.squeeze()
print(f'x_squeezed tensor shape {x_squeezed.shape}')
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f'x_unsqueezed tensor shape {x_unsqueezed.shape}')
```

```python
#torch.permute - rearranges the dimensions of a target tensor in a specified order
# perfmute is a
x_original = torch.rand(size=(224, 224, 3))
print(f'x_original tensor shape {x_original.shape}')
x_permuted = x_original.permute(2, 0, 1)
print(f'x_permuted tensor shape {x_permuted.shape}')
```

### Pytorch Reproducability (trying to take random out of random.

In Short how a neural network learns.

`start with random numbers` -> tensor Operations -> updata random numbers to try and make them better representations of the data.->again -> agaom.

To reduce the randomness in neural networks and PyTorch comes the concept of a random seed.

Essentially what the random seed does is falvour the randomness.


```python
import torch
# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Lets nake some random but reproducible tensors

import torch
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED)   #we have to reset the random seed.
random_tensor_D = torch.rand(3, 4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
