### Creating a range of tensors and tensors like (takes the shape and creates those many zeroes or ones)

```python
one_to_thousand = torch.arange(start=1, end = 1000, step = 88)
one_to_thousand
```

### Tensor Datatypes
*** Note : ** Tensor datatypes in one of the 3 big errors you will run in PyTorch and DeepLearning

1. Tensors not right data type use tensor.dtype
2. Tensors not right shape use tensor.shape
3. Tensors not on the right device. use tensor.device

```python
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               device=None,
                               requires_grad=False,
                               dtype=None) # What dataype is the tensor. check torch.tensor in the google for datatypes.

float_32_tensor
float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor.dtype
```

```python
some_tensor = torch.rand(3, 4)

some_tensor.dtype, some_tensor.shape, some_tensor.device
```

### Matrix Multiplication

Two main ways of performing multiplication in neural networks and deep learning

1.   Element Wise Multiplication
2.   Matrix Multiplication (dot product)



```python
#1 Element wise Mutliplication

tensor * tensor
```

```python
%%time

value = 0
for i in range(len(tensor)):
  value += tensor[i]*tensor[i]
print(value)
```

```python
%%time

torch.matmul(tensor, tensor)
```

### one of the most common error is shape error.
### Two main rules that performing matrix multiplication needs to satisfy

1. inner dimensions must match
2. resulting matrix has the shape of the ** outer dimensions **

```python
torch.matmul(torch.rand(2, 10), torch.rand(10, 2))
```
